```python
import pandas as pd
import numpy as np
import pickle
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
```

```python
with open("pipeline_preprocessing.pkl", "rb") as file:
    pipeline_preprocessing  = pickle.load(file )
```

```python
df_train = pd.read_csv('data/train.csv')
```

```python
df_train_proc = pd.DataFrame(
    pipeline_preprocessing.fit_transform(df_train), 
    columns=pipeline_preprocessing['transformer_preprocessing'].get_feature_names_out()
)

df_train_proc.head()
```

```python
X = df_train
y = df_train.sales
```

```python
pipeline = Pipeline([
    ('pipeline_preprocessing', pipeline_preprocessing),
    ('linear_regression', LinearRegression())
])
```

```python
pipeline.fit(X,y)
```

```python
y_pred = pipeline.predict(X)
```

```python
y_pred.size
```

```python
z = np.zeros(y_pred.size)
z
```

```python
a = np.vstack((y_pred, z))
```

```python
a
```

```python
b = np.amax(a,0)
```

```python
serie_b = pd.Series(b+1)
serie_b
```

```python
np.log(serie_b/(y+1))
```

```python
serie_b/np.log(y+1)
```
